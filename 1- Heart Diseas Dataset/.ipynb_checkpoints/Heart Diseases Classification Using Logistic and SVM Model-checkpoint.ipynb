```python
x = df.iloc[:,0:13]
```

```python
y = df.iloc[:,13]
```

```python
y_2 = y.values
```

```python
y_22 = y_2.reshape(-1,1)
```

# Standard Scaler convert values

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
data_y = sc.fit_transform(y_22)
data_x = sc.fit_transform(x)
```

# PCA Analysis

```python
from sklearn.decomposition import PCA
pca = PCA(n_components = 4)
x_data = pca.fit_transform(data_x)
explain_variance = pca.explained_variance_ratio_
```

```python
explain_variance
```

# Train Test Split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(x_data,y_22,test_size=0.2,random_state=0)
```

# Apply SVM Using Radial Basis Function

```python
from sklearn import svm
from sklearn import metrics
clf = svm.SVC(kernel = 'rbf')
clf.fit(X_train,Y_train)
pred = clf.predict(X_test)
```

```python
print("Accuracy",metrics.accuracy_score(Y_test,pred))
```

# Confusion Matrix

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,pred)
```

```python
cm
```

# Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
clas = LogisticRegression(random_state = 0)
clas.fit(X_train,Y_train)
l_predict = clas.predict(X_test)
```

```python
print("Accuracy",metrics.accuracy_score(Y_test,l_predict))
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test,l_predict)
```
